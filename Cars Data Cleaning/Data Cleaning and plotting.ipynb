```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data=pd.read_csv('Example.csv')
raw_data.describe(include='all')
```

```python
raw_data.isnull().sum()
```

```python
data_no_nan=raw_data.dropna(axis=0)
```

```python
data_no_nan.describe(include='all')
```

```python
data_no_nan=data_no_nan.drop(['Model'],axis=1)
```

```python
data_no_nan
```

```python
#creating the distribution graph of column variables
#Price
sns.displot(data_no_nan['Price'])
```

```python
#We can see that there some outliers on the right of the graph so we need to remove it for better readability
#Removing outliers from 'Price'data
q=data_no_nan['Price'].quantile(0.99)
data_no_nan=data_no_nan[data_no_nan['Price']<q]
sns.displot(data_no_nan['Price'])
```

```python
#We'll do the same things with 'Mileage' 'EngineV' and 'Yaer'
```

```python
sns.displot(data_no_nan['Year'])
```

```python
#We have outliers in the left in this case
q1=data_no_nan['Year'].quantile(0.01)
data_no_nan=data_no_nan[data_no_nan['Year']>q1]
sns.displot(data_no_nan['Year'])
```

```python
#Mileage case
sns.displot(data_no_nan['Mileage'])
```

```python
q2=data_no_nan['Mileage'].quantile(0.99)
data_no_nan=data_no_nan[data_no_nan['Mileage']<q2]
sns.displot(data_no_nan['Mileage'])
```

```python
#EngineV case
sns.displot(data_no_nan['EngineV'])
```

```python
q3=data_no_nan['EngineV'].quantile(0.99)
data_no_nan=data_no_nan[data_no_nan['EngineV']<q3]
sns.displot(data_no_nan['EngineV'])
```

```python
data_no_nan
```

```python
#After dropping some rows we can notice that the order of indexing is missing the order (0 1 3 4 7...)
#We need to reset the indexing of the dataframe
data_cleaned=data_no_nan.reset_index(drop=True)
data_cleaned
```

```python
data_cleaned.describe(include='all')
```

```python
#saving the cleaned_data to csv file
data_cleaned.to_csv('Cars Data.csv',index=False)
```
